```python
import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
```

```python
nltk.download('punkt')
```

```python
medical_text = """ Diabetes is a chronic disease that affects how the body processes blood sugar.
If untreated, diabetes may cause heart disease, kidney failure, nerve damage and vision problems.
Early diagnosis and proper treatment help improve patient outcomes.
"""
```

```python
nltk.download('punkt_tab')
sentences = sent_tokenize(medical_text)
print("Sentence Tokenization:\n")
for s in sentences:
    print("-", s)
```

```python
words = word_tokenize(medical_text)
print("Word Tokenization:\n")
print(words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in words if word.isalpha()]

print("Stemmed Words:\n")
print(stemmed_words)
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(medical_text)

lemmatized_words = [token.lemma_ for token in doc if token.is_alpha]

print("Lemmatized Words:\n")
print(lemmatized_words)
```

```python
comparison = list(zip(stemmed_words[:15], lemmatized_words[:15]))
comparison
```
